Ejemplo Bioinformático de busqueda por metodo de fuerza bruta


```python
def find (target, sequence): 
    for i in range(len(sequence)):
        counter = 0
        
        for j in range(i, min(i + len(target), len(sequence))):
            if target[j - i] == sequence[j]:
                counter += 1
        
            if counter == len(target):
                return (i, i + len(target) - 1)
    
    return (-1, -1)

target = 'AAGGCAGATGGGCTATATAAACGTTTTCGCTT'
(ini, fin) = find(target, sequence)

assert(sequence.find(target) == ini)

print('Search  {}'.format(target))
print('Found   {} (indices from {} to {})'.format(sequence[ini: fin + 1], ini, fin))
```
